```python
import tkinter
import datetime
from tkinter import *

root = Tk()
root.title("calculator")
root.geometry("500x300")
root.config(background="lightblue")

l1=Label(text="Enter Ur name Plz : ")
l1.pack()

e1=Entry()
e1.pack()

l2=Label(text="Enter Ur BD date Plz : ")
l2.pack()

# e2=Entry()
# e2.pack()

l3=Label(text="Enter the year Plz : ")
l3.pack()

e3=Entry()
e3.pack()

l4=Label(text="Enter the monthe Plz : ")
l4.pack()

e4=Entry()
e4.pack()

l5=Label(text="Enter the day Plz : ")
l5.pack()

e5=Entry()
e5.pack()

def calc ():
    y = int(e3.get())
    m = int(e4.get())
    d = int(e5.get())
    n = e1.get()

    bd = datetime.datetime(y,m,d)
#   print(f"Ur BD is {bd}")

    cy = datetime.datetime.now().year
    cm = datetime.datetime.now().month
    cd = datetime.datetime.now().day

    hly = cy - y 
    hlm = cm - m
    hld = d - cd
    
    resultlbl=Label(text=(f"Hi {n}\nU Live For {hly} Year , And {hlm} Month  , And {hld} Days "))
    resultlbl.pack()
#   print(f"U Live For {hly} Year , And {hlm} Month  , And {hld} Days ")

    
b1=Button(text="clic" , command=calc ,bg="pink" , font="30")
b1.pack()


root.mainloop()
```
